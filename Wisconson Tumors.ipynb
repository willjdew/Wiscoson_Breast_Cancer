```python
def count_true_false_predictions(test, predictions):
    # set the parameters to 0
    true_negatives = 0
    false_negatives = 0
    true_positives = 0
    false_positives = 0
    
    # for each prediction compare to truth and count and see if they are the 
    # same or different.
    for prediction, truth in zip(predictions, test):
        if prediction == 0 and truth == 0:
            true_negatives += 1
        elif prediction == 0 and truth == 1:
            false_negatives += 1
        elif prediction == 1 and truth == 0:
            false_positives += 1
        elif prediction == 1 and truth == 1:
            true_positives += 1
        else:
            print "Warning: Found a predicted label not == 0 or 1."
            print "All predictions should take value 0 or 1."
            print "Evaluating performance for processed predictions:"
            return False
        
    # create a dictionary with all the parameters
    counts = {'true_negatives': true_negatives,
              'false_negatives': false_negatives,
              'true_positives': true_positives,
              'false_positives': false_positives}
    
    return counts
```

```python
def accuracy_precision_recall_f1(counts, clf):
    # extract all the parameters
    true_negatives = counts['true_negatives']
    false_negatives = counts['false_negatives']
    true_positives = counts['true_positives']
    false_positives = counts['false_positives']
    
    try:
        # Calculations for accuracy, precision, recall, and f1
        total_predictions = true_negatives + false_negatives + false_positives + true_positives
        accuracy = 1.0*(true_positives + true_negatives)/total_predictions
        precision = 1.0*true_positives/(true_positives+false_positives)
        recall = 1.0*true_positives/(true_positives+false_negatives)
        f1 = 2.0 * true_positives/(2*true_positives + false_positives+false_negatives)
        f2 = (1+2.0*2.0) * precision*recall/(4*precision + recall)
        peram = {"Accuracy": accuracy, "Precision": precision, "Recall": recall, "F1": f1}
        return peram
    except:
        print "Got a divide by zero when trying out clf:", clf
        print "Precision or recall may be undefined due to a lack of true positive predicitons."
```

```python
def clf_predictions(clf, data, labels, outcome):
    # split our data into train and test
    train, test = train_test_split(data, test_size = 0.3)
    
    # split train, test data into x and y
    train_X = train[labels]
    train_y = train[outcome]
    test_X = test[labels]
    test_y = test[outcome]
    
    # now fit our model for traiing data
    clf.fit(train_X, train_y)
    
    # predict for the test data
    predictions = clf.predict(test_X)
    acc = metrics.accuracy_score(predictions, test_y)
    
    # find the true, false preditions
    test_y_np = test_y.values
    counts = count_true_false_predictions(test_y_np, predictions)
    
    # find accuracy, precision, recall, and f1
    if counts:
        peram = accuracy_precision_recall_f1(counts, clf)
        print clf
        print peram
```

```python
# https://www.analyticsvidhya.com/blog/2018/05/improve-model-performance-cross-validation-in-python-r/
def classification_model(clf, data, labels, outcome, folds=1000):
    # create our X, y data for kfolds
    X = data[labels].copy()
    y = data[outcome]
    
    # using Stratified k-fold cross validation
    sss = StratifiedShuffleSplit(n_splits=folds, test_size=0.5, random_state=0)
    
    # holds each k-fold model's estimation error
    error = []
    
    # split our data into train and test
    for train_index, test_index in sss.split(X, y):
        # split train, test data into x and y
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        # now fit our model for traiing data
        clf.fit(X_train, y_train)
    
        error.append(clf.score(X_test, y_test))
        
    # returns the mean of the clf estimations errors
    print 'estimations error (model bias): ', np.mean(error)
    print 'standard deviation (model variance):', np.std(error)
```

```python
model_rfc = RandomForestClassifier(n_estimators=100)
classification_model(model_rfc, data, features_mean, 'diagnosis', 5)
```

```python
model_svm = svm.SVC()
classification_model(model_svm, data, features_mean, 'diagnosis', 5)
```

```python
model_knc = KNeighborsClassifier()
classification_model(model_knc, data, features_mean, 'diagnosis', 5)
```

```python
model_lr = LogisticRegression()
classification_model(model_lr, data, features_mean, 'diagnosis', 5)
```
